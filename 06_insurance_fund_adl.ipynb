```python
import pandas as pd
import numpy as np
```

```python
n = 100

df = pd.DataFrame({

    "account_id": range(n),

    "symbol": np.random.choice(
        ["BTC", "ETH"],
        n
    ),

    "price": np.random.randint(
        10000,
        30000,
        n
    ),

    "position": np.random.randint(
        1,
        50,
        n
    ),

    "balance": np.random.randint(
        1000,
        20000,
        n
    ),

    "leverage": np.random.choice(
        [5, 10, 20, 50],
        n
    )

})
```

```python
df["notional"] = df["price"] * df["position"]
```

```python
df["risk_ratio"] = (
    df["notional"] / df["balance"]
)
```

```python
df["pnl"] = np.random.randn(len(df)) * 1000
```

```python
df["adl_score"] = (
    df["pnl"]
    * df["risk_ratio"]
    * df["leverage"]
)
```

```python
adl_queue = df.sort_values(
    "adl_score",
    ascending=False
)

adl_queue.head(10)
```

```python
insurance = 1_000_000

loss = 2_000_000

adl_trigger = loss > insurance

adl_trigger
```

```python
if adl_trigger:

    print("ADL triggered")

    print(
        adl_queue.head(5)
    )
```

```python
btc_queue = df[
    df["symbol"] == "BTC"
].sort_values(
    "adl_score",
    ascending=False
)

btc_queue.head()
```

```python
df["adl_rank"] = pd.qcut(
    df["adl_score"],
    5,
    labels=False
)

df.head()
```

```python
print("Max score:",
      df["adl_score"].max())

print("Top ADL:",
      adl_queue.head(3))
```
